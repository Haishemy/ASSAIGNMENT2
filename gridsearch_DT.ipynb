```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid= {'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
             'splitter':['best', 'random'],
             'max_features':['sqrt','log2']}
grid= GridSearchCV (DecisionTreeRegressor(), param_grid, refit=True, verbose=0, n_jobs=1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("the best r2 value is:{}".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
import pickle
filename="finalized_model_DT.sav"
pickle.dump(grid,open(filename,'wb'))
loaded_model=pickle.load(open("finalized_model_DT.sav",'rb'))
result=loaded_model.predict([[33,22,3,1,0]])
```
